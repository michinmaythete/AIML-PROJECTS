```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten

```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train #data is in pixel form 3-D array
```

```python
x_train.shape #28by28 array
```

```python
#pixel to image
import matplotlib.pyplot as plt    #first image
plt.imshow(x_train[0])
```

```python
#bring all values inbetween 0 to 1
#divide every element in matrix by 255
x_train=x_train/255
x_test=x_test/255
```

```python
x_test
```

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28))) #convert 3D to 1D
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy'\
              ,optimizer='Adam',metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,epochs=25,validation_split=0.2)
```

```python
y_prob=model.predict(x_test)
```

```python
y_prob #here we got the prob of digits
```

```python
y_pred=y_prob.argmax(axis=1)   #builtin function
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','validation'],loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','validation'],loc='upper left')
plt.show()
```
